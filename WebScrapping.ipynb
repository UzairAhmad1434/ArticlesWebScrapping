```python
import urllib.request
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
```

```python
data = pd.read_csv(r'U:\ML\WebScrapping\data.csv')
```

```python
data
```

```python
data.columns = ['urls']
title = []
sub_title=[]
article=[]
author=[] 
author_urls=[] 
clap=[] 
read_time=[] 
image_source=[]
data_urls_scraped=[]
articles=[]


```

```python
for i in range(500):
    try:
        url_to_scrap = data.iloc[i]['urls']
        # print(url_to_scrap)
        r = requests.get(url_to_scrap, timeout=10)
        text_got = r.text+''
        r.close()
        data_urls_scraped.append(url_to_scrap)
        
        soup = BeautifulSoup(text_got, "html.parser")
        
        h1_tag = soup.find('h1')
        title = h1_tag.text.strip()

        h2_h3_tags = soup.find_all(['h2', 'h3'])  
        sub_title = [h.text.strip() for h in h2_h3_tags]
        sub_title = [subtitle for subtitle in sub_title if not subtitle.startswith("Responses (")]


        import re
        buttons = soup.select("button.af, button.ag, button.ah, button.ai, button.aj, button.ak, button.al, button.am, button.an, button.ao, button.ap, button.aq, button.ar, button.as, button.at, button.aih, button.ne")
        if buttons:
            for button in buttons:
                button_text = button.text.strip()
                if re.search(r'\d', button_text): 
                    clap = button_text
                    break  
        else:
            print("No buttons found")




        read_time_span = soup.find("span", {"data-testid": "storyReadTime"})
        if read_time_span:
            read_time = read_time_span.text.strip()


        authorName=soup.find("a",{"data-testid": "authorName"})
        if authorName:
            author=authorName.text.strip()

        text_blocks = soup.find_all("p", class_="pw-post-body-paragraph")  
        text_blocks = text_blocks[:5]

        article = "\n".join(p.get_text(strip=True) for p in text_blocks)  

        images = soup.find_all("img") 
        image_source = [img["src"] for img in images if img.get("src")] 

        author_link = soup.find("a", {"data-testid": "authorName"}) 
        if author_link and author_link.get("href"):
            author_urls = author_link["href"]

            
        dict = {
            "url": url_to_scrap,
            "title": title,  
            "sub_title": sub_title,
            "claps":clap,
            "read_time":read_time,
            "author":author,
            "article":article,
            "image_source":image_source,
            "author_urls":author_urls
        }

        articles.append(dict)

    except:
        print("url not found")

```

```python
article_data=pd.DataFrame(articles)
```
